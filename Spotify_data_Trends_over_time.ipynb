```
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
%config inlineBackend.figure_format='retina'
plt.xkcd()
```

```
date_data = pd.read_csv('data_by_year.csv')
```

```
date_data.head()
```

```
date_data.info()
```

```
date_data.describe()
```

```
list(date_data.columns)
```

```
uniform_data = date_data[['acousticness',
 'danceability',
 'duration_ms',
 'energy',
 'instrumentalness',
 'liveness',
 'loudness',
 'speechiness',
 'tempo',
 'valence',
 'popularity',
 'key']].corr()
```

```
fig,ax= plt.subplots(figsize=(10,7))
sns.heatmap(uniform_data)
```

```
fig,ax= plt.subplots(figsize=(15,7))
sns.lineplot(x="year", y="acousticness", data=date_data)
```

```
fig,ax= plt.subplots(figsize=(15,7))
sns.lineplot(x="year", y="energy", data=date_data, ax=ax)
```

```
fig,ax= plt.subplots(figsize=(15,7))
sns.lineplot(x="year", y="duration_ms", data=date_data, ax=ax)
```

```
plt.figure(figsize=(16,10))
sns.set(style="whitegrid")
columns = ["danceability","energy"]
for col in columns:
  x = date_data.groupby("year")[col].mean()
  ax= sns.lineplot(x=x.index, y=x, label=col)
ax.set_title("danceability energy relationship over the years")  
ax.set_ylabel('Measure')
ax.set_xlabel('year')
```
